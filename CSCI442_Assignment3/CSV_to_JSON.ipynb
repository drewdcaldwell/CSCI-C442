<h2>CSV to JSON converter using Pandas Library</h2>

<P>Using the pandas library we want to convert the twitter.csv file from a CSV to a formatted JSON without changing any of the schema. Using the Pandas library makes this simple.</P>

```python
# Import the pandas library
import pandas as pd
```

```python
# Intitialize the CSV file you want to convert.
csv_file = 'twitter.csv'        #### IF DIFFERENT CSV CHANGE HERE <---------####
df = pd.read_csv(csv_file)

# Output the first few rows to see what data we have.
df.head()
```

```python
# Use the 'to_json' function to get the dataframe to convert it's values into JSON format.
json_output = df.to_json(orient = "records", indent = 4)
# Output what we have for the JSON format
print(json_output)
```

```python
json_file = "CSV_to_JSON_output.json"
with open(json_file, "w") as f:
    f.write(json_output)

print(f"JSON file saved successfully as (json_file)")
```
